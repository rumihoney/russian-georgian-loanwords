# Russian Loanwords in Georgian

This notebook analyzes a manually annotated dataset of Russian loanwords in the Georgian language. 

```python
from russianloans import data
```

```python
# Checking dataset size
len(data)
```

```python
# Count categories 
category_counts = {}

for word, info in data.items():
    category = info["category"].strip().lower()

    if category == "":
        category = "unclassified"

    if category not in category_counts:
        category_counts[category] = 0

    category_counts[category] += 1

category_counts
```

```python
# Results neatly formatted
for category, count in sorted(category_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{category}: {count}")
```

```python
# Percentages

total = sum(category_counts.values())

for category, count in sorted(category_counts.items(), key=lambda x: x[1], reverse=True):
    percent = (count / total) * 100
    print(f"{category}: {percent:.2f}%")
```

## Interpretation

The results show which semantic categories are most represented in the dataset.
These patterns can be discussed in relation to Russian-Georgian contact and historical influence.
