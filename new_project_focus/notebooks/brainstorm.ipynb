```python
import json
from collections import Counter

with open("../data/geo_borrowed_from_russian.json", encoding="utf-8") as f:
    loanwords = json.load(f)

print("Total number of loanwords:", len(loanwords))
```

```python
loanwords[:10]
```

```python
word_lengths = [len(word) for word in loanwords]

print("Shortest word length:", min(word_lengths))
print("Longest word length:", max(word_lengths))
print("Average word length:", round(sum(word_lengths) / len(word_lengths), 2))
```

```python
first_letters = [word[0] for word in loanwords]
first_letter_counts = Counter(first_letters)

first_letter_counts.most_common(10)
```

```python
final_letters = [word[-1] for word in loanwords]
final_letter_counts = Counter(final_letters)

final_letter_counts.most_common(10)
```

```python
georgian_vowels = ("ა", "ე", "ი", "ო", "უ")

vowel_final_words = [word for word in loanwords if word[-1] in georgian_vowels]
consonant_final_words = [word for word in loanwords if word[-1] not in georgian_vowels]

print("Vowel-final words:", len(vowel_final_words))
print("Consonant-final words:", len(consonant_final_words))
```

```python
target_endings = ["ი", "ა", "ობა", "აცია", "იზმი", "ენტი", "ორი"]

for ending in target_endings:
    matching_words = [word for word in loanwords if word.endswith(ending)]
    print(ending, ":", len(matching_words))
```
